```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
prior = [0.728, 0.891, 0.707, 0.766, 0.756, 0.792]
prior_no_images = [2418, 1507, 2402, 2417, 2418, 2365]
prior_labels = ['RDSNet', 'S4Net', 'BASNet', 'CPD-R', 'SCRN', 'Siris']
```

```python
new = [0.654, 0.718]
new_no_images = [2347, 2347, 2347, 2347, 2347]
new_labels = ['Original SaRa', 'SaRa + $\epsilon$ + $\sigma$ + $\gamma$']
```

```python
all = prior + new
all_no_images = prior_no_images + new_no_images
all_labels = prior_labels + new_labels

all = np.array(all)
```

```python
pc = '#6096b3'
nc = '#7d7d7d'
```

```python
colors = [pc, pc, pc, pc, pc, pc, '#4a4a4a', '#bf4e4e']
```

```python
plt.figure(figsize=(12, 8))
plt.bar(np.arange(len(all)), all, color=colors)

plt.xlabel('Model', fontsize=24)
plt.ylabel('SOR', fontsize=24)

# Add padding to ylabel
plt.gca().xaxis.labelpad = 1
plt.gca().yaxis.labelpad = 20

# Add horizontal lines at the tops of the bars which span the entire plot
for i in range(len(all)):
    if i > 5:
        plt.axhline(y=all[i] - 0.002, color=colors[i], linestyle='--', linewidth=2)

# Change opacities of bars according to all_no_images
# for i in range(len(all)):
#     plt.gca().patches[i].set_alpha(all_no_images[i] / 2418)

# Plot a line at y = 0.765
plt.axhline(y=0.765 - 0.002, color='orange', linestyle='--', linewidth=2, label='Mean SOR')

plt.xticks(np.arange(len(all)), all_labels, rotation=45, fontsize=14, ha='right', rotation_mode='anchor')

# Set last tick label to be bold
labels = plt.gca().get_xticklabels()
labels[-1].set_fontweight('bold')

plt.yticks(np.arange(0, 1.1, 0.1), fontsize=14)
plt.ylim(0, 1)
plt.legend(fontsize=16)
# plt.grid(axis='y', color='#808080', linewidth=0.5)
plt.show()
```
